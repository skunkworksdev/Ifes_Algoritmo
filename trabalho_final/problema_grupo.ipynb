```python
m = 0.00028
R = 0.14
g = 9.8
```

```python
# Letra A: Encontre a altura h para que a bola perca o contato com o trilho
# no ponto mais alto do circulo.

h = (((20 * R * g) + (7 * R * g)) / (10 * g))
print(h)
```

```python
# Letra B: Se agora h = 6R, encontre o vetor força que o trilho faz sobre
# a bola no ponto Q?

N = m * (((50 * R * g) / 7) / R)
print(N)
```
